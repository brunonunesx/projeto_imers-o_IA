```python
# Passo 1 Importar a base de dado
import pandas as pd

tabela = pd.read_csv("clientes.csv") # importa a base de dados
display(tabela)

#Good -> Boa
#Standard -> OK
#Poor -> Ruim
```

É importante salientar que a base de dados utilizada é fictícia e foi elaborada exclusivamente para fins de teste e validação do modelo.

```python
#Passo 2 - Preparar a base de daados para a inteligência artificial
# verificar se temos valores vazios ou valores reconhecidos em formato errado
print(tabela.info())
print(tabela.columns)
```

Após importar a base de dados fictícia, a primeira etapa é verificar a existência de valores vazios ou formatação incorreta, pois podem prejudicar a análise. Neste caso, a base de dados está pronta para uso, sem necessidade de tratamento prévio.
É crucial observar que algumas colunas possuem o tipo "object", representando informações textuais. Modelos de classificação não processam textos diretamente, sendo necessário realizar um tratamento nessas colunas antes de prosseguir com o projeto.

Tirando a coluna score_credito, que é a que a inteligencia artificial vai prever, todas as outra devem ser númericas

então devemos nesse caso tratar a colunas
profissão
mix_credito
comportamento_ pagamento

pq estão no formato object.

```python
from sklearn.preprocessing import LabelEncoder
# vai transformar as colunas de texto em números, ex: profissoes vai sair de cientista, professor, mecanico, etc para 0, 1, 2, etc
codificador = LabelEncoder()

# só não aplicamos na coluna de score_credito que é o nosso objetivo
for coluna in tabela.columns:
    if tabela[coluna].dtype == "object" and coluna != "score_credito":
        tabela[coluna] = codificador.fit_transform(tabela[coluna])

# verificando se realmente todas as colunas foram modificadas
print(tabela.info())
```

```python
# escolhendo quais colunas vamos usar para treinar o modelo
# y é a coluna que queremos que o modelo calcule
# x vai todas as colunas que vamos usar para prever o score de credito, não vamos usar a coluna id_cliente porque ela é um numero qualquer que nao ajuda a previsao
x = tabela.drop(["score_credito", "id_cliente"], axis=1)
y = tabela["score_credito"]

from sklearn.model_selection import train_test_split

# separamos os dados em treino e teste. Treino vamos dar para os modelos aprenderem e teste vamos usar para ver se o modelo aprendeu corretamente
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
```

Aplicando o conhecimento adquirido na segunda aula do projeto Imersão IA, treinei as IAs utilizando a técnica de Few-Shot Prompting. Essa técnica consiste em fornecer à IA alguns exemplos específicos da tarefa que se deseja realizar, permitindo que ela aprenda a partir desses exemplos e realize a tarefa solicitada.

```python
# 3 Passo - Criar o modelo de IA -> Prever o score e Credito: Bom, Ok, Ruim
#Importat a ineligencia artificial
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
# criar a intelligencia artificial
modelo_arvore = RandomForestClassifier() # modelo arvore de decisao
modelo_knn = KNeighborsClassifier() # modelo do KNN (nearest neighbors - vizinhos mais proximos)
# treinando os modelos
modelo_arvore.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

Tendo em vista que o Gemini ainda não se encontrava disponível para uso deste metodo, a alternativa encontrada foi utilizar duas outras Inteligências Artificiais, provenientes de um projeto anterior, com as quais já possuía familiaridade.

```python
#4 Passo - Escolher qual o melhor modelo de IA
#acurácia = accuracy
from sklearn.metrics import accuracy_score

# calculamos as previsoes
previsao_arvore = modelo_arvore.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

# comparamos as previsoes com o y_teste
# esse score queremos o maior (maior acuracia, mas tb tem que ser maior do que o chute de tudo Standard)
print(accuracy_score(y_teste, previsao_arvore))
print(accuracy_score(y_teste, previsao_knn))
```

explicação do que é acurácia no readme do projeto mas basicamente
acurácia é a avaliação de modelos de classificação. A acurácia mede a performance do modelo, indicando a porcentagem de classificações corretas.
nesse casso o modelo_arvore foi o melhor então e o que irei utlizar

```python
#5 Passo - Usar a noossa IA para definir o sccore de crédito dos clientes
#aqui tambem precisamos trasformar as colunas de textos dos novos clientes.
novos_clientes = pd.read_csv("novos_clientes.csv")
display(novos_clientes)
for coluna in novos_clientes.columns:
    if novos_clientes[coluna].dtype == "object" and coluna != "score_credito":
        novos_clientes[coluna] = codificador.fit_transform(novos_clientes[coluna])

previsoes = modelo_arvore.predict(novos_clientes)
print(previsoes)
```
